```python
df['precision'] = 100*(df['goals']/df['ontarget_scoring_att'])
```

```python
df['chance_out'] = 100*(df['total_yel_card']/df['total_tackle'])
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='precision',y='wins')
```

```python
x = np.array(df['precision']).reshape(-1,1)
y = np.array(df['wins']).reshape(-1,1)

model = LinearRegression()
model.fit(x,y)
```

```python
a = model.coef_[0][0]
b = model.intercept_[0]
a,b
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df,x='precision',y='wins')
plt.plot(x, a*x + b, c='red')
```
